```r
library(ggplot2)
options(repr.plot.width=4, repr.plot.height=3)
```

```r
this_model <- function(x, y=1) {
    if (y == 1)
        return(dnorm(x, mean=0, sd=1))
    else if (y == 2)
        return(dnorm(x, mean=1, sd=1e6))
}
```

A. Find the decision boundary

```r
idx <- seq(-6, 6, length.out=1000)
df <- data.frame(i = idx,
                 class_1 = sapply(idx, this_model, y=1),
                 class_2 = sapply(idx, this_model, y=2))

ggplot(df, aes(x=idx)) +
    geom_line(aes(y=class_1, color='red')) +
    geom_line(aes(y=class_2, color='blue')) +
    scale_y_log10()

# Find the roots
decision.point <- df[(df$class_1 - df$class_2) == min(df$class_1 - df$class_2), ]$i
print(paste0('Decision boundary occurs at: ', decision.point))
```

B. Now suppose sigma_2 = 1. What does the decision boundary look like now?

```r
idx <- seq(-6, 6, length.out=1000)
df.2 <- data.frame(i = idx,
                   class_1 = dnorm(idx, mean=0, sd=1),
                   class_2 = dnorm(idx, mean=1, sd=1))

ggplot(df.2, aes(x=idx)) +
    geom_line(aes(y=class_1, color='red')) +
    geom_line(aes(y=class_2, color='blue')) +
    scale_y_log10()

# Find the roots
decision.point <- df.2[(df.2$class_1 - df.2$class_2) == min(df.2$class_1 - df.2$class_2), ]$i
print(paste0('Decision boundary occurs at: ', decision.point))
```
